## Sets

A set is a collection which is unordered, mutable and does not allow duplicates.

Note: Set elements must be of an immutable type.

```python
#defining a set:
my_set = {}
my_set = set()


my_set = {1,2,3,10,12,15}
my_set = set([1,2,3,10,12,15])
my_set = set((1,2,3,10,12,15))
print(my_set)
```

```python
str_set = set("Banana")
print(str_set)
```

Sets are unorderd and unindexed, so you can't access a item in a set by index.

```python
#adding an item to the a set:
my_set.add(5)
my_set.add(6)
print(my_set)
```

```python
#update a set with another set:
s1 = {'a', 'b', 'c'}
s2 = {'d', 'e', 'f','a'}

s1.update(s2)
print(s1)
```

```python
#removing an item from a set:
my_set.remove(5)
print(my_set)
```

```python
my_set.remove(7)
```

```python
my_set.discard(7)
```

```python
#removing an item randomly from a set:
x = my_set.pop()
print(x)
```

```python
#clears a set
my_set.clear()
```

```python
#checking membership in a set:
5 in my_set
```

```python
2 in my_set
```
